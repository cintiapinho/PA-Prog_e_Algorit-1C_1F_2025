Media notas

solicite 3 notas e realize a media

image.png

```python
# entrada

nota1 = float(input("Digite a nota1: "))
nota2 = float(input("Digite a nota1: "))
nota3 = float(input("Digite a nota1: "))

#processamento

media = (nota1+nota2+nota3)/3

#saida

print("A média das notas é: ", media)
print(f"A média das notas é: {media}")
```
